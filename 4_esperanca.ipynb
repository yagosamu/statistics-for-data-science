# Esperança de uma variável aleatória

Exemplo: Seja X o valor que sai na face superior de um dado. Calcule o valor esperado de X.

```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

medias = []
S = range(1, 1000)  # número de simulações

for s in S:  # simula
    saidas_dado = []
    for n in range(0, s):  # pequenas simulações
        # guarda o valor na face do dado
        saidas_dado.append(randint(1, 6))
    
    # armazena o valor esperado
    medias.append(np.mean(saidas_dado))

# mostra os resultados
plt.figure(figsize=(8, 6))
plt.plot(S, medias)
plt.axhline(3.5, color='red', linestyle='--')  # valor esperado teórico
plt.xlabel('Número de lançamentos')
plt.ylabel('Média')
plt.title('Lei dos Grandes Números - Dado')
plt.grid()
plt.show()
```

**Exemplo:** Uma padaria produz pães que proporcionam um lucro de R\\$1,00 real, quando vendidos, ou um prejuízo de R\$0,50 quando são encaminhados para virarem farinha, no outro dia.  Se a probabilidade de vender um pão é igual a 0,8, qual o lucro médio da padaria?


```python
import numpy as np

Ns = 1000 # número de simulações
p = 0.8 # probabilidade de vender um pão
lucro = 1 # lucro por pão
prejuizo = 0.5 # prejuízo por pão
ganho = 0 # lucro médio
for s in range(0,Ns):
    # se vende um pão
    if np.random.uniform() < p:     #Aqui o raciocínio é imaginar uma reta de 0 a 1, onde 0.8 
        ganho = ganho + lucro           #é o ponto de corte para vender um pão. Se o número gerado 
    # caso contrário                    #for menor que 0.8 consideramos que o pão foi vendido.
    else:
        ganho = ganho - prejuizo
print('Lucro médio: ', ganho/Ns)
```

**Exemplo:** Lançamos dois dados e observamos a variável aleatória
X , que é igual a 1 se o valor observado na soma de dois dados for
par, ou igual a zero, caso contrário. Calcule a esperança de
g (X ) = X + 2.

Valor calculado na aula: E[g(X)] = 2,5

```python
import numpy as np

# define a função g(X)
def g(x):
  x = x + 2
  return x

dado = np.array([1,2,3,4,5,6])
ns = 100 # número de simulações
vg = [] # lista que guarda os valores de g
for i in range(ns):
  x = 0
  dado1 = np.random.choice(dado)
  dado2 = np.random.choice(dado)
  soma = dado1 + dado2
  if soma % 2 == 0: # se a soma for par
    x = 1
  else: # se a soma for ímpar
    x = 0
  vg.append(g(x))
print('Esperança de g(X): ', np.mean(vg))
```
