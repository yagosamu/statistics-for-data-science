#### Probabilidade Condicional e Teorema de Bayes

**Probabilidade Condicional**

A probabilidade de um evento A ocorrer dado que B já ocorreu:

P(A|B) = P(A ∩ B) / P(B)

---

**Teorema de Bayes**

Permite inverter probabilidades condicionais:

P(A|B) = (P(B|A) * P(A)) / P(B)

**Probabilidade condicional simples**

Exemplo: Probabilidade de chover (A) dado que está nublado (B)

```python
P_A = 0.3      # P(chover)
P_B = 0.5      # P(nublado)
P_AeB = 0.2    # P(chover e nublado)

P_A_dado_B = P_AeB / P_B
P_A_dado_B
```

**Exemplo: Teorema de Bayes (teste médico)**

```python
P_doenca = 0.01
P_pos_dado_doenca = 0.95
P_pos = 0.10

P_doenca_dado_pos = (P_pos_dado_doenca * P_doenca) / P_pos

print(f"P(doença | positivo) = {P_doenca_dado_pos:.4f}")
```

**Simulação**

```python
import numpy as np

n = 10000

doenca = np.random.rand(n) < 0.01
teste = np.where(doenca, np.random.rand(n) < 0.95, np.random.rand(n) < 0.05)

prob = doenca[teste].mean()

print(f"Probabilidade estimada: {prob:.4f}")
```

**Insight**

Mesmo com um teste preciso, a probabilidade real de ter a doença dado um resultado positivo pode ser baixa quando a doença é rara.

Isso mostra a importância do Teorema de Bayes na prática.

```python
import matplotlib.pyplot as plt

# Quantidade de positivos
positivos = teste

# Verdadeiros positivos (tem doença e teste positivo)
verdadeiros = doenca & teste

# Falsos positivos (não tem doença mas teste positivo)
falsos = (~doenca) & teste

valores = [verdadeiros.sum(), falsos.sum()]
labels = ["Verdadeiros Positivos", "Falsos Positivos"]

plt.figure()
plt.bar(labels, valores)
plt.title("Resultados do Teste")
plt.ylabel("Quantidade")
plt.xticks(rotation=20)

plt.show()
```
