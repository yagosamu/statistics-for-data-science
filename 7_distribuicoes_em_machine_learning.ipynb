#### Estatística → Machine Learning

Este notebook mostra como distribuições estatísticas se conectam com modelos de Machine Learning.

A ideia não é focar nas fórmulas, mas entender como cada modelo aparece na prática.

### Distribuição Normal → Regressão Linear

A regressão linear assume que os erros seguem uma distribuição normal.

Isso permite:
- estimar coeficientes
- calcular intervalos de confiança
- fazer inferência estatística

```python
import numpy as np
import matplotlib.pyplot as plt

# simulando dados
np.random.seed(42)
x = np.linspace(0, 10, 100)
erro = np.random.normal(0, 1, 100)
y = 2*x + 3 + erro

# gráfico
plt.scatter(x, y)
plt.title("Regressão Linear com Erro Normal")
plt.show()
```

### Distribuição Binomial → Classificação

A binomial modela número de sucessos (0 ou 1).

Base de:
- classificação binária
- regressão logística

```python
from scipy.stats import binom

# probabilidade de 7 sucessos em 10 tentativas
p = binom.pmf(7, n=10, p=0.5)
p
```

### Distribuição de Bernoulli → Regressão Logística

Cada observação em classificação binária é um Bernoulli:

- 1 = sucesso
- 0 = fracasso

A regressão logística modela P(Y=1).

```python
from scipy.special import expit  # função sigmoid

x = np.linspace(-5, 5, 100)
y = expit(x)

plt.plot(x, y)
plt.title("Função Sigmoid (Regressão Logística)")
plt.show()
```

### Distribuição de Poisson → Contagem de eventos

Usada quando queremos prever quantidade de eventos:

- número de acessos
- chamadas
- ocorrências

Base de modelos como Poisson Regression

```python
data = np.random.poisson(lam=3, size=1000)

plt.hist(data, bins=20)
plt.title("Distribuição de Poisson")
plt.show()
```

### Distribuição Exponencial → Tempo entre eventos

Modela o tempo entre ocorrências.

Muito usada em:
- sistemas de fila
- análise de sobrevivência

```python
data = np.random.exponential(scale=1/2, size=1000)

plt.hist(data, bins=30)
plt.title("Distribuição Exponencial")
plt.show()
```

### Distribuição Beta → Probabilidades (Bayesian)

Modela probabilidades entre 0 e 1.

Muito usada em:
- testes A/B
- inferência Bayesiana

```python
from scipy.stats import beta

x = np.linspace(0, 1, 100)
y = beta.pdf(x, a=2, b=5)

plt.plot(x, y)
plt.title("Distribuição Beta")
plt.show()
```

### Distribuição t-Student → Amostras pequenas

Usada quando temos poucos dados.

Importante para:
- intervalos de confiança
- testes estatísticos

```python
from scipy.stats import t

x = np.linspace(-4, 4, 100)
y = t.pdf(x, df=5)

plt.plot(x, y)
plt.title("Distribuição t-Student")
plt.show()
```

### Insight Final

Distribuições estatísticas não são apenas teoria.

Elas são a base de muitos modelos de Machine Learning:

- Normal → erros e regressão
- Binomial/Bernoulli → classificação
- Poisson → contagem
- Exponencial → tempo entre eventos
- Beta → probabilidades
- t-Student → inferência com poucos dados

Entender essas conexões permite aplicar modelos de forma mais consciente e eficiente.
