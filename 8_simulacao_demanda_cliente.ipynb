#### Simulação de Demanda de Clientes com Modelos Estatísticos

Este projeto simula o comportamento de demanda de clientes utilizando distribuições estatísticas e conecta esses conceitos a problemas reais de negócio.

### Objetivo

Simular:

- Número de pedidos (Poisson)
- Tempo entre pedidos (Exponencial)
- Conversão de clientes (Binomial)
- Valor dos pedidos (Normal)

Objetivo final: entender demanda, receita e risco operacional.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

### Pedidos por hora (Distribuição de Poisson)

Assumimos uma média de 8 pedidos por hora.

```python
pedidos_por_hora = np.random.poisson(lam=8, size=24)

plt.figure()
plt.plot(pedidos_por_hora, marker='o')
plt.title("Pedidos por Hora")
plt.xlabel("Hora")
plt.ylabel("Quantidade de Pedidos")

plt.show()
```

### Tempo entre pedidos (Distribuição Exponencial)

Modela o tempo até o próximo pedido ocorrer.

```python
tempo_entre_pedidos = np.random.exponential(scale=1/8, size=1000)

plt.figure()
plt.hist(tempo_entre_pedidos, bins=30)
plt.title("Tempo entre Pedidos")

plt.show()
```

### Conversão de clientes (Distribuição Binomial)

Cada cliente pode:
- 1 = comprar
- 0 = não comprar

Assumimos uma taxa de conversão de 12%.

```python
conversao = np.random.binomial(1, 0.12, size=1000)

taxa_conversao = conversao.mean()

print(f"Taxa de conversão estimada: {taxa_conversao:.2f}")
```

### Valor dos pedidos (Distribuição Normal)

Assumimos:
- Média = 120
- Desvio padrão = 30

```python
valores_pedidos = np.random.normal(120, 30, size=1000)

plt.figure()
plt.hist(valores_pedidos, bins=30)
plt.title("Distribuição do Valor dos Pedidos")

plt.show()
```

### Simulação de Receita

```python
receita = valores_pedidos * conversao

receita_total = receita.sum()

print(f"Receita total simulada: {receita_total:.2f}")
```

### Insights

- A demanda varia ao longo do tempo (Poisson)
- A maioria dos pedidos ocorre em intervalos curtos (Exponencial)
- A taxa de conversão apresenta incerteza (Binomial)
- A receita varia de acordo com o comportamento dos clientes (Normal)
